```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
nelx, nely = 60, 20         # número de elementos en x e y
volfrac = 0.4               # fracción de volumen permitida
penal = 3.0                 # penalización SIMP
rmin = 1.5                  # radio mínimo de filtro
E0 = 1.0                    # módulo de Young del material
Emin = 1e-9                 # mínimo valor para evitar singularidades

# Inicialización del campo de densidad (material)
x = volfrac * np.ones((nely, nelx))

# Visualización inicial
def plot_density(x, title="Distribución de material"):
    plt.figure(figsize=(10, 4))
    plt.imshow(x, cmap='gray', origin='lower')
    plt.colorbar(label="Densidad")
    plt.title(title)
    plt.xlabel("Elementos en X")
    plt.ylabel("Elementos en Y")
    plt.show()

plot_density(x)
```
